```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('Frequency')
plt.title('Histogram for SAT SCORE')
plt.show()
```

```python
sns.histplot(df1['SAT'])
plt.xlabel('SATSCORES')
plt.ylabel('Frequency')
plt.title('Histogram for SAT SCORE')
plt.show()
```

```python
sns.histplot(df1['Expenses'],color='brown')
plt.xlabel('Expenses')
plt.ylabel('Frequency')
plt.title('Histogram for Expenses')
plt.show()
```
